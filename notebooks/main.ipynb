## Load and Inspect Employee Data

This cell imports the main analysis libraries, loads the employee dataset from `data/raw/Employee_Data.csv`, and shows its shape and first few rows for a quick sanity check.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set_style('whitegrid')

df = pd.read_csv('../data/raw/Employee_Data.csv')
print(f"Shape: {df.shape}")
df.head(n=10)
```

## Cell 4 — Missing values


```python
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_summary = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
missing_summary[missing_summary['missing_count'] > 0].sort_values('missing_pct', ascending=False)
```

## Target variable distribution (Performance_Score)

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

df['Performance_Score'].value_counts().sort_index().plot(kind='bar', ax=axes[0])
axes[0].set_title('Performance_Score Distribution (counts)')
axes[0].set_xlabel('Performance Score')

sns.histplot(df['Performance_Score'], bins=5, kde=True, ax=axes[1])
axes[1].set_title('Performance_Score Distribution (density)')

plt.tight_layout()
plt.show()

print(df['Performance_Score'].value_counts(normalize=True).sort_index())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

df['Department'].value_counts().plot(kind='barh', ax=axes[0])
axes[0].set_title('Employees per Department')

df['Job_Title'].value_counts().plot(kind='barh', ax=axes[1])
axes[1].set_title('Employees per Job Title')

plt.tight_layout()
plt.show()
```

```python
dept_perf = df.groupby('Department')['Performance_Score'].agg(['mean', 'std', 'count']).sort_values('mean', ascending=False)
print(dept_perf)

plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Department', y='Performance_Score')
plt.xticks(rotation=45)
plt.title('Performance_Score by Department')
plt.tight_layout()
plt.show()
```

```python
numeric_cols = df.select_dtypes(include='number').columns
corr = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Matrix (numeric features)')
plt.tight_layout()
plt.show()

# Specifically check correlation with target
print(corr['Performance_Score'].sort_values(ascending=False))
```

```python
print("Correlation of Monthly_Salary with Performance_Score:",
      df['Monthly_Salary'].corr(df['Performance_Score']))

sns.scatterplot(data=df, x='Monthly_Salary', y='Performance_Score', alpha=0.1)
plt.title('Monthly_Salary vs Performance_Score')
plt.show()
```

```python
outcome_like = ['Projects_Handled', 'Promotions', 'Sick_Days']
for col in outcome_like:
    print(f"{col} vs Performance_Score correlation:", df[col].corr(df['Performance_Score']))
```
